```python
import time
def source2dest(L, destStart, sourceStart, range):
    
    output = []
    for i,l in enumerate(L):
        if sourceStart <= int(l) < sourceStart+range:
            output.append([i,destStart+int(l)-sourceStart])
    return output
```

```python
def fusion(L):
    for i,x in enumerate(L):
        for j,y in enumerate(L):
            if x[1]==y[0]:
                L[i]= [x[0],y[1]]
                L.remove([y[0],y[1]])
    return L

def update(new, old, s_o, s_n, r):
    cop = old.copy()
    for interval in cop:
        si,ei = interval[0], interval[1]
        if si >= s_o:
            #...-----...
            #..-----
            if s_o <= si <= s_o+r-1 and ei >= s_o+r:
                new.append([s_n+si-s_o, s_n+r-1])
                old.remove([si,ei])
                old.append([s_o+r,ei])
            #...-----...
            #..-------..
            elif ei <= s_o + r - 1:
                new.append([s_n+si-s_o, s_n+ei-s_o])
                old.remove([si,ei])

        elif si <= s_o:
            #...-----..
            #....-----.
            if s_o <= ei <= s_o + r-1:
                new.append([s_n, s_n+ei-s_o])
                old.remove([si,ei])
                old.append([si,s_o-1])
            #...-----..
            #.....--...
            elif ei >= s_o+r-1:
                new.append([s_n,s_n+r-1])
                old.remove([si,ei])
                old.append([si,s_o-1])
                old.append([s_o+r,ei])


    return new, old
```
