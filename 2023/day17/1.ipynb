```python
import numpy as np
```

```python
grid = []
with open('input.txt') as f:
    for line in f.readlines():
        l = []
        for c in line.strip():
            l.append(int(c))
        grid.append(l)
grid
```

```python
A = np.array(grid) #the weight of the graph
G = [(i,j) for i in range(A.shape[0]) for j in range(A.shape[1])] #the vertices of the graph
D = {}  #the distance from (0,0) to any vertex
for s in G:
    D[s] = np.inf
D[(0,0)] = 0

P=[]    #le graphe vide
```

```python
def find_min(Q,D):
    mini = np.inf
    sommet = (-1,-1)
    for vertex in Q:
        if D[vertex] < mini:
            mini = D[vertex]
            sommet = vertex
    return sommet

def maj_dist(s1, s2, D, pred):
    if D[s2]>D[s1]+A[s2[0],s2[1]]:
        D[s2]=D[s1]+A[s2[0],s2[1]]
        if s1 in pred.keys():
            pred[s2] = pred[s1] + [np.array(s1)]
        else : 
            pred[s2] = [np.array(s1)]
    return D, pred

def neigh(s, pred, m, n):
    next_s = []
    if s in pred.keys():
        pred_s = pred[s]
        print(f'predecessors: {pred_s}')
        dir = s - pred_s[-1]

        # + or - 90 degrees turns
        dn, dm = dir[0], dir[1]
        if 0<=s[0]+dm<n and 0<=s[1]+dn<m:
            next_s.append((s[0]+dm,s[1]+dn))
        if 0<=s[0]-dm<n and 0<=s[1]-dn<m:
            next_s.append((s[0]-dm,s[1]-dn))

        if len(pred_s)<3:
            if 0<=s[0]+dn<n and 0<=s[1]+dm<m:
                next_s.append((s[0]+dn,s[1]+dm))
        else:
            if np.array_equal(pred_s[-2], s-2*dir) and np.array_equal(pred_s[-3], s-3*dir):
                print("End of straight...")
            else:
                if 0<=s[0]+dn<n and 0<=s[1]+dm<m:
                    next_s.append((s[0]+dn,s[1]+dm))
    else:
        if 0<=s[0]+1<n :
            next_s.append((s[0]+1,s[1]))
        if 0<=s[0]-1<n:
            next_s.append((s[0]-1,s[1]))
        if 0<=s[1]+1<m:
            next_s.append((s[0],s[1]+1))
        if 0<=s[1]-1<m:
            next_s.append((s[0],s[1]-1))
        
    
    return next_s

def Dijkstra(A, sdeb):
    n, m = A.shape[0], A.shape[1]
    G = [(i,j) for i in range(n) for j in range(m)] #the vertices of the graph
    D = {}  #the distance from (0,0) to any vertex
    for s in G:
        D[s] = np.inf
    D[sdeb] = 0
    pred = {}

    while G!=[]:
        s1 = find_min(G,D)
        print(f'current vortex : {s1}')
        G.remove(s1)
        for s2 in neigh(s1, pred, m, n):
            print(s2)
            D, pred = maj_dist(s1, s2, D, pred)
        print(G)
        print(D)
    return D,pred
```
