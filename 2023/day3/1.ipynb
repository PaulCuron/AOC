```python
with open('input.txt') as f:
    M = []
    for line in f.readlines():
        M.append(line.split()[0])
```

```python
def get_len(M, i, j):
    if M[i][j]=='*':
        return 1
    else:
        size = 1
        while (j+size < len(M[0])) and (M[i][j+size].isdigit()):
            size+=1
        sizeb = 0
        while (j-sizeb-1 >= 0) and (M[i][j-sizeb-1].isdigit()):
            sizeb+=1 
        return size+sizeb

def get_complen(M, i, j):
    size = 0
    while (j+size < len(M[0])) and (M[i][j+size].isdigit()):
        size+=1
    sizeb = 0
    while (j-(sizeb+1) >= 0) and (M[i][j-(sizeb+1)].isdigit()):
        sizeb+=1 
    return [size,sizeb]

def get_neigh(M, i, j):
    h = len(M)
    w = len(M[0])
    size = get_len(M,i,j)
    neigh = []
    if i > 0:
        
        if j > 0:
            neigh.append([i-1,j-1])
        for k in range(size):
            neigh.append([i-1,j+k])
        if j+size < w:
            neigh.append([i-1,j+size])

    if j > 0:
            neigh.append([i,j-1])
    if j+size < w:
        neigh.append([i,j+size])

    if i < h - 1:
        
        if j > 0:
            neigh.append([i+1,j-1])
        for k in range(size):
            neigh.append([i+1,j+k])
        if j+size < w:
            neigh.append([i+1,j+size])
    return neigh

def isPart(M, i, j):
    neigh = get_neigh(M,i,j)
    for pos in neigh:
        c = M[pos[0]][pos[1]]
        if (not c.isdigit()) and (c !='.'):
            return True
    return False
```

```python
i,j = 0,0
h,w = len(M), len(M[0])
Sum = 0
while i < h:
    j = 0
    while j < w:
        size = get_len(M,i,j)

        if M[i][j].isdigit():
            if isPart(M,i,j):
                Sum += int(M[i][j:j+size])
            j += size
        else:
            j+=1
    i+=1
```

```python
def isGear(M,i,j):
    if M[i][j]!='*':
        return 0
    else:
        neighs = get_neigh(M,i,j)
        k = 0
        ints = []
        while k < len(neighs):
            nb = neighs[k]
            if M[nb[0]][nb[1]].isdigit():
                size = get_complen(M,nb[0],nb[1])
                k += 1
                n = int(M[nb[0]][nb[1]-size[1]:nb[1]+size[0]])
                if not n in ints:
                    ints.append(n)
            else:
                k += 1
        if len(ints) == 2:
            return ints[0]*ints[1]
        else:
            return 0
```

```python

h,w = len(M), len(M[0])
Sum = 0
for i in range(h):
    for j in range(w):
        Sum+=isGear(M,i,j)
```
